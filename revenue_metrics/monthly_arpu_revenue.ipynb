```python
"""
Monthly revenue + users + ARPU (practice)

Goal:
- compute total revenue, unique users
- aggregate monthly revenue and users
- calculate ARPU by month

Note: learning notebook
"""

import pandas as pd

# 1) Load data
df = pd.read_csv("/content/orders_practice_dataset.csv")  # <- replace if needed
df.head()
```

```python
# 2) Parse dates + keep only paid orders
df["date"] = pd.to_datetime(df["order_date"])
df = df[df["revenue"] > 0].copy()

# Quick totals
total_revenue = df["revenue"].sum()
users = df["user_id"].nunique()
orders = len(df)

arpu_total = total_revenue / users

total_revenue, users, orders, arpu_total
```

```python
# 3) Monthly aggregation
df["month"] = df["date"].dt.to_period("M")

monthly = (
    df.groupby("month")
      .agg(
          revenue=("revenue", "sum"),
          users=("user_id", "nunique"),
          orders=("user_id", "size")
      )
      .sort_index()
)

monthly["ARPU"] = monthly["revenue"] / monthly["users"]
monthly
```
